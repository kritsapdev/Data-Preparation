```python
import pandas as pd
import numpy as np
```

```python
# load dataset
df = pd.read_csv("/content/custinfo.csv")
df.head(10)
```

```python
replacement_char = ''
# Apply replacement using regular expressions
df = df.replace(to_replace='"', value=replacement_char, regex=True)
df = df.replace(to_replace='ok', value=replacement_char, regex=True)
df = df.replace(to_replace=' ', value=replacement_char, regex=True)
df = df.replace(to_replace='NaN', value=replacement_char, regex=True)

# Function to impute missing values using mode for a single column
def impute_missing_mode(col):
  mode_value = df[col].mode()[0]  # mode is get the most frequent value
  df[col].fillna(mode_value, inplace=True)

# Loop through columns and impute missing values
for col in df.columns:
  impute_missing_mode(col)
# to check missing values
missing_values = df.isnull().sum()
missing_values

df.head(10)
```
